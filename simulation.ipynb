```python
# 6b
# After 100 students, what is chance of at least one corrupted install?

n_trials = n_rows = 1_000_000

## Install Conda n times  (100)

n_installs = n_cols = 100

# Corrupted or Clean Install
p_corrupt = 0.0004

# generate random values betweeen 0 and 1
data = np.random.random((n_rows,n_cols))
data

#compare those random values with 0.0004
corrupted = data < p_corrupt
corrupted

# create dataframe using bools instead of the randomly generated values
corrupted_bool100_df = pd.DataFrame(corrupted)
corrupted_bool100_df

# create a calculated column to sum the total number of True
corrupted_bool100_df['corrupted'] = corrupted_bool100_df.iloc[:,0:50].sum(axis=1)
corrupted_bool100_df

# find mean value
corrupted100_pct = corrupted_bool100_df.corrupted.mean()
corrupted100_pct

# again, subtract the answer the specific question
```

```python
# 7a Given a 70% chance of FOOD TRUCK, how (un)likely is it to NOT see FOOD TRUCK for the last 3 days?

## set the number of trials:

n_trials = n_rows = 1_000_000

## set number of data points (3 days)

n_days = n_cols = 3

# Food_Truck or NO
p_food = 0.7
```
